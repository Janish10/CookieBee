````python
# Cell 3: Enhanced Cookie Analysis Engine with AI-Powered Insights
# ================================================================
class EnhancedCookieAnalysisEngine:
    def __init__(self):
        self.known_trackers = {
            'google_analytics': {
                'cookies': ['_ga', '_gid', '_gat', '__utma', '__utmb', '__utmc', '__utmz'],
                'company': 'Google',
                'purpose': 'Website Analytics',
                'risk_level': 'Medium'
            },
            'facebook_pixel': {
                'cookies': ['_fbp', '_fbc', 'fr', 'datr', 'sb', 'wd'],
                'company': 'Meta (Facebook)',
                'purpose': 'Social Media Tracking',
                'risk_level': 'High'
            },
            'doubleclick': {
                'cookies': ['IDE', 'DSID', 'FLC', 'GAPS', 'GALX'],
                'company': 'Google',
                'purpose': 'Advertising',
                'risk_level': 'High'
            },
            'hotjar': {
                'cookies': ['_hjid', '_hjSessionUser', '_hjSession', '_hjIncludedInPageviewSample'],
                'company': 'Hotjar',
                'purpose': 'User Behavior Analytics',
                'risk_level': 'Medium'
            },
            'amplitude': {
                'cookies': ['amplitude_id', 'amplitude_sessionId'],
                'company': 'Amplitude',
                'purpose': 'Product Analytics',
                'risk_level': 'Medium'
            }
        }

        self.risk_categories = {
            'ESSENTIAL': {
                'score': 1,
                'color': '#4CAF50',
                'icon': '‚úÖ',
                'description': 'Required for basic website functionality'
            },
            'FUNCTIONAL': {
                'score': 2,
                'color': '#2196F3',
                'icon': 'üîß',
                'description': 'Enhances user experience and preferences'
            },
            'ANALYTICAL': {
                'score': 3,
                'color': '#FF9800',
                'icon': 'üìä',
                'description': 'Tracks user behavior and website performance'
            },
            'MARKETING': {
                'score': 4,
                'color': '#F44336',
                'icon': 'üéØ',
                'description': 'Used for advertising and cross-site tracking'
            }
        }

        # Enhanced EdTech platform data with more varied cookie profiles
        self.edtech_platforms = {
            'khanacademy.org': {
                'name': 'Khan Academy',
                'category': 'K-12 Education',
                'cookies': [
                    {'name': '_ga', 'domain': '.khanacademy.org', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Track learning progress', 'size': '24 bytes'},
                    {'name': '_gid', 'domain': '.khanacademy.org', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Session tracking', 'size': '16 bytes'},
                    {'name': 'session_id', 'domain': 'khanacademy.org', 'category': 'ESSENTIAL', 'purpose': 'User authentication and session management', 'size': '32 bytes'},
                    {'name': 'csrf_token', 'domain': 'khanacademy.org', 'category': 'ESSENTIAL', 'purpose': 'Cross-site request forgery protection', 'size': '28 bytes'},
                    {'name': 'user_prefs', 'domain': 'khanacademy.org', 'category': 'FUNCTIONAL', 'purpose': 'Store user interface preferences', 'size': '64 bytes'},
                    {'name': '_fbp', 'domain': '.khanacademy.org', 'category': 'MARKETING', 'purpose': 'Facebook Pixel - Track donations and engagement', 'size': '36 bytes'},
                    {'name': 'hotjar_id', 'domain': '.khanacademy.org', 'category': 'ANALYTICAL', 'purpose': 'Hotjar - Analyze user interaction patterns', 'size': '20 bytes'},
                    {'name': 'learn_progress', 'domain': 'khanacademy.org', 'category': 'FUNCTIONAL', 'purpose': 'Track student learning progress', 'size': '128 bytes'}
                ]
            },
            'coursera.org': {
                'name': 'Coursera',
                'category': 'Higher Education & Courses',
                'cookies': [
                    {'name': '_ga', 'domain': '.coursera.org', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Course engagement tracking', 'size': '24 bytes'},
                    {'name': '_gid', 'domain': '.coursera.org', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Session behavior', 'size': '16 bytes'},
                    {'name': 'CAUTH', 'domain': 'coursera.org', 'category': 'ESSENTIAL', 'purpose': 'User authentication token', 'size': '64 bytes'},
                    {'name': '__stripe_mid', 'domain': '.coursera.org', 'category': 'FUNCTIONAL', 'purpose': 'Stripe payment processing', 'size': '32 bytes'},
                    {'name': '_fbp', 'domain': '.coursera.org', 'category': 'MARKETING', 'purpose': 'Facebook advertising for course promotion', 'size': '36 bytes'},
                    {'name': 'IDE', 'domain': '.doubleclick.net', 'category': 'MARKETING', 'purpose': 'DoubleClick advertising network', 'size': '28 bytes'},
                    {'name': 'optimizelyEndUserId', 'domain': '.coursera.org', 'category': 'ANALYTICAL', 'purpose': 'A/B testing for course recommendations', 'size': '40 bytes'},
                    {'name': '_hjid', 'domain': '.coursera.org', 'category': 'ANALYTICAL', 'purpose': 'Hotjar user behavior tracking', 'size': '20 bytes'},
                    {'name': 'course_progress', 'domain': 'coursera.org', 'category': 'FUNCTIONAL', 'purpose': 'Track course completion status', 'size': '256 bytes'},
                    {'name': 'user_segments', 'domain': 'coursera.org', 'category': 'MARKETING', 'purpose': 'Segment users for targeted ads', 'size': '50 bytes'} # Added variation
                ]
            },
            'udemy.com': {
                'name': 'Udemy',
                'category': 'Online Learning Marketplace',
                'cookies': [
                    {'name': '_ga', 'domain': '.udemy.com', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Student behavior tracking', 'size': '24 bytes'},
                    {'name': 'access_token', 'domain': 'udemy.com', 'category': 'ESSENTIAL', 'purpose': 'API access authentication', 'size': '128 bytes'},
                    {'name': 'client_id', 'domain': 'udemy.com', 'category': 'ESSENTIAL', 'purpose': 'Client session identification', 'size': '32 bytes'},
                    {'name': 'seen_cookie_banner', 'domain': 'udemy.com', 'category': 'FUNCTIONAL', 'purpose': 'Track cookie consent acknowledgment', 'size': '8 bytes'},
                    {'name': '_fbp', 'domain': '.udemy.com', 'category': 'MARKETING', 'purpose': 'Facebook Pixel for course advertising', 'size': '36 bytes'},
                    {'name': '_gcl_au', 'domain': '.udemy.com', 'category': 'MARKETING', 'purpose': 'Google Ads conversion tracking', 'size': '32 bytes'},
                    {'name': 'OptanonConsent', 'domain': 'udemy.com', 'category': 'FUNCTIONAL', 'purpose': 'OneTrust consent management', 'size': '512 bytes'},
                    {'name': 'learning_path', 'domain': 'udemy.com', 'category': 'FUNCTIONAL', 'purpose': 'Personalized learning recommendations', 'size': '192 bytes'},
                     {'name': 'ad_user_id', 'domain': '.doubleclick.net', 'category': 'MARKETING', 'purpose': 'Ad tracking user identifier', 'size': '40 bytes'}, # Added variation
                     {'name': 'affiliate_id', 'domain': 'udemy.com', 'category': 'MARKETING', 'purpose': 'Affiliate marketing tracking', 'size': '20 bytes'} # Added variation
                ]
            },
            'edx.org': {
                'name': 'edX',
                'category': 'University-level MOOCs',
                'cookies': [
                    {'name': 'sessionid', 'domain': '.edx.org', 'category': 'ESSENTIAL', 'purpose': 'Django session management', 'size': '64 bytes'},
                    {'name': 'csrftoken', 'domain': '.edx.org', 'category': 'ESSENTIAL', 'purpose': 'CSRF protection for forms', 'size': '32 bytes'},
                    {'name': '_ga', 'domain': '.edx.org', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Course engagement', 'size': '24 bytes'},
                    {'name': '_gid', 'domain': '.edx.org', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Learning analytics', 'size': '16 bytes'},
                    {'name': 'edx-user-info', 'domain': '.edx.org', 'category': 'FUNCTIONAL', 'purpose': 'User profile and preferences', 'size': '256 bytes'},
                    {'name': '_fbp', 'domain': '.edx.org', 'category': 'MARKETING', 'purpose': 'Facebook tracking for university partnerships', 'size': '36 bytes'},
                    {'name': 'course_enrollment', 'domain': 'edx.org', 'category': 'FUNCTIONAL', 'purpose': 'Track enrolled courses', 'size': '128 bytes'},
                    {'name': 'partner_tracking', 'domain': '.edx.org', 'category': 'ANALYTICAL', 'purpose': 'University partner analytics', 'size': '30 bytes'} # Added variation
                ]
            },
            'duolingo.com': {
                'name': 'Duolingo',
                'category': 'Language Learning',
                'cookies': [
                    {'name': 'session', 'domain': 'duolingo.com', 'category': 'ESSENTIAL', 'purpose': 'User session and login state', 'size': '48 bytes'},
                    {'name': '_ga', 'domain': '.duolingo.com', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Learning progress tracking', 'size': '24 bytes'},
                    {'name': '_gid', 'domain': '.duolingo.com', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics - Daily active users', 'size': '16 bytes'},
                    {'name': 'auth_tkt', 'domain': 'duolingo.com', 'category': 'ESSENTIAL', 'purpose': 'Authentication ticket for API access', 'size': '72 bytes'},
                    {'name': '_fbp', 'domain': '.duolingo.com', 'category': 'MARKETING', 'purpose': 'Facebook Pixel for app downloads', 'size': '36 bytes'},
                    {'name': 'amplitude_id', 'domain': '.duolingo.com', 'category': 'ANALYTICAL', 'purpose': 'Amplitude analytics - Learning patterns', 'size': '40 bytes'},
                    {'name': 'streak_data', 'domain': 'duolingo.com', 'category': 'FUNCTIONAL', 'purpose': 'Track daily learning streaks', 'size': '64 bytes'},
                    {'name': 'language_prefs', 'domain': 'duolingo.com', 'category': 'FUNCTIONAL', 'purpose': 'Language learning preferences', 'size': '96 bytes'},
                    {'name': 'ad_conversion', 'domain': '.duolingo.com', 'category': 'MARKETING', 'purpose': 'Ad conversion tracking', 'size': '30 bytes'} # Added variation
                ]
            },
             'grammarly.com': { # Added new platform
                'name': 'Grammarly',
                'category': 'Educational Tools',
                'cookies': [
                    {'name': 'client_session', 'domain': 'grammarly.com', 'category': 'ESSENTIAL', 'purpose': 'Session management', 'size': '50 bytes'},
                    {'name': '_ga', 'domain': '.grammarly.com', 'category': 'ANALYTICAL', 'purpose': 'Usage analytics', 'size': '24 bytes'},
                    {'name': '_fbp', 'domain': '.grammarly.com', 'category': 'MARKETING', 'purpose': 'Advertising', 'size': '36 bytes'},
                    {'name': 'user_settings', 'domain': 'grammarly.com', 'category': 'FUNCTIONAL', 'purpose': 'User preferences', 'size': '100 bytes'},
                    {'name': 'IDE', 'domain': '.doubleclick.net', 'category': 'MARKETING', 'purpose': 'Advertising network', 'size': '28 bytes'},
                    {'name': 'feature_flags', 'domain': 'grammarly.com', 'category': 'FUNCTIONAL', 'purpose': 'Feature rollout tracking', 'size': '80 bytes'}
                ]
            },
            'codepen.io': { # Added new platform
                'name': 'CodePen',
                'category': 'Educational Tools',
                'cookies': [
                    {'name': 'session', 'domain': 'codepen.io', 'category': 'ESSENTIAL', 'purpose': 'User session', 'size': '60 bytes'},
                    {'name': '_ga', 'domain': '.codepen.io', 'category': 'ANALYTICAL', 'purpose': 'Usage analytics', 'size': '24 bytes'},
                    {'name': '_fbp', 'domain': '.codepen.io', 'category': 'MARKETING', 'purpose': 'Advertising', 'size': '36 bytes'},
                    {'name': 'user_preferences', 'domain': 'codepen.io', 'category': 'FUNCTIONAL', 'purpose': 'Editor settings', 'size': '150 bytes'},
                    {'name': 'hotjar_id', 'domain': '.codepen.io', 'category': 'ANALYTICAL', 'purpose': 'User behavior', 'size': '20 bytes'},
                    {'name': 'ad_id', 'domain': '.doubleclick.net', 'category': 'MARKETING', 'purpose': 'Ad personalization', 'size': '30 bytes'}
                ]
            },
             'notion.so': { # Added new platform
                'name': 'Notion',
                'category': 'Educational Workspace',
                'cookies': [
                    {'name': 'token', 'domain': 'notion.so', 'category': 'ESSENTIAL', 'purpose': 'Authentication', 'size': '150 bytes'},
                    {'name': '_ga', 'domain': '.notion.so', 'category': 'ANALYTICAL', 'purpose': 'Workspace analytics', 'size': '24 bytes'},
                    {'name': 'client_state', 'domain': 'notion.so', 'category': 'FUNCTIONAL', 'purpose': 'App state', 'size': '200 bytes'},
                     {'name': 'session_id', 'domain': 'notion.so', 'category': 'ESSENTIAL', 'purpose': 'Session management', 'size': '40 bytes'}
                ]
            }
        }


    def simulate_advanced_cookie_scan(self, url):
        """Enhanced cookie scanning with detailed analysis - Dynamically generate or use known data"""
        domain = urlparse(url).netloc.lower().replace('www.', '')

        # Get platform-specific data or dynamically generate
        platform_data = self.edtech_platforms.get(domain)

        if platform_data:
             cookies_to_analyze = platform_data['cookies']
             platform_name = platform_data['name']
             platform_category = platform_data['category']
        else:
            # Simulate dynamic cookie detection for unknown platforms
            # Introduce more variability based on domain structure or type
            simulated_cookies = [
                {'name': 'session_id', 'domain': domain, 'category': 'ESSENTIAL', 'purpose': 'User session', 'size': f'{np.random.randint(30, 80)} bytes'},
                {'name': '_ga', 'domain': f'.{domain}', 'category': 'ANALYTICAL', 'purpose': 'Google Analytics', 'size': f'{np.random.randint(20, 30)} bytes'},
                {'name': 'user_preferences', 'domain': domain, 'category': 'FUNCTIONAL', 'purpose': 'Site preferences', 'size': f'{np.random.randint(50, 150)} bytes'},
            ]

            # Randomly add marketing/third-party cookies based on heuristics
            if 'edu' not in domain: # Less marketing for .edu (simulated)
                 if np.random.rand() > 0.3: # 70% chance of marketing cookies
                    simulated_cookies.append({'name': '_fbp', 'domain': f'.{domain}', 'category': 'MARKETING', 'purpose': 'Facebook Pixel', 'size': f'{np.random.randint(30, 40)} bytes'})
                 if np.random.rand() > 0.4: # 60% chance of third-party ad cookie
                     simulated_cookies.append({'name': 'IDE', 'domain': '.doubleclick.net', 'category': 'MARKETING', 'purpose': 'Advertising', 'size': f'{np.random.randint(25, 35)} bytes'})
                 if np.random.rand() > 0.6: # 40% chance of additional analytics
                     simulated_cookies.append({'name': 'amplitude_id', 'domain': f'.{domain}', 'category': 'ANALYTICAL', 'purpose': 'Amplitude Analytics', 'size': f'{np.random.randint(35, 45)} bytes'})

            # Add some random essential/functional cookies
            for i in range(np.random.randint(1, 4)):
                 simulated_cookies.append({'name': f'misc_{i}', 'domain': domain, 'category': np.random.choice(['ESSENTIAL', 'FUNCTIONAL']), 'purpose': 'Miscellaneous', 'size': f'{np.random.randint(10, 100)} bytes'})

            cookies_to_analyze = simulated_cookies
            platform_name = domain.replace('.com', '').replace('.org', '').replace('.io', '').replace('.so', '').replace('.edu', '').title()
            platform_category = 'Simulated EdTech'


        # Add timestamp and scanning metadata
        scan_data = {
            'url': url,
            'domain': domain,
            'platform_name': platform_name,
            'platform_category': platform_category,
            'timestamp': datetime.now().isoformat(),
            'cookies': cookies_to_analyze,
            'total_cookies': len(cookies_to_analyze),
            'scan_duration': np.random.uniform(3.0, 6.0),  # Slightly longer simulated scan time
            'response_time': np.random.uniform(1.0, 3.0)   # Slightly longer simulated response time
        }

        return scan_data

    def analyze_cookies_advanced(self, scan_data):
        """Enhanced cookie analysis with detailed insights - Emphasize impact of specific cookies"""
        cookies = scan_data['cookies']

        # Initialize analysis metrics
        category_counts = {}
        risk_score = 0
        cookie_details = []
        total_size = 0
        tracker_companies = set()

        for cookie in cookies:
            category = cookie['category']
            category_counts[category] = category_counts.get(category, 0) + 1
            risk_score += self.risk_categories[category]['score'] # Keep base risk score from categories

            # Extract size in bytes
            size_str = cookie.get('size', '0 bytes')
            size_bytes = int(re.findall(r'\d+', size_str)[0]) if re.findall(r'\d+', size_str) else 0
            total_size += size_bytes

            # Identify tracker companies
            is_third_party = cookie['domain'].startswith('.') or urlparse(cookie['domain']).netloc != scan_data['domain']
            tracker = 'Unknown'
            for tracker_name, tracker_info in self.known_trackers.items():
                if cookie['name'] in tracker_info['cookies']:
                    tracker = tracker_info['company']
                    tracker_companies.add(tracker_info['company'])
                    break # Found a match, move to next cookie

            cookie_details.append({
                'name': cookie['name'],
                'domain': cookie['domain'],
                'category': category,
                'purpose': cookie['purpose'],
                'size': size_str,
                'size_bytes': size_bytes,
                'risk_level': self.get_risk_level(category),
                'third_party': is_third_party, # More accurate third-party check
                'icon': self.risk_categories[category]['icon'],
                'tracker_company': tracker # Add identified tracker company
            })

        # Calculate enhanced privacy metrics - MORE SENSITIVITY TO COOKIE TYPES AND COUNT
        total_cookies = len(cookies)
        marketing_count = category_counts.get('MARKETING', 0)
        analytical_count = category_counts.get('ANALYTICAL', 0)
        third_party_count = sum(1 for c in cookie_details if c['third_party'])

        # Adjust risk score based on specific cookie types and counts
        adjusted_risk_score = (
            (marketing_count * 5) + # Higher weight for marketing
            (analytical_count * 3) + # Medium weight for analytical
            (third_party_count * 2) + # Additional weight for third-party
            (total_cookies * 0.5)   # Small penalty for total count
        )

        # Normalize score
        max_possible_adjusted_risk = (total_cookies * 5) # Assuming max possible is all marketing third-party
        privacy_score = max(0, 100 - (adjusted_risk_score / max_possible_adjusted_risk * 100) if max_possible_adjusted_risk > 0 else 100)

        # Calculate additional metrics
        third_party_ratio = third_party_count / total_cookies * 100 if total_cookies > 0 else 0
        marketing_ratio = marketing_count / total_cookies * 100 if total_cookies > 0 else 0


        return {
            'total_cookies': total_cookies,
            'category_counts': category_counts,
            'privacy_score': round(privacy_score, 1),
            'privacy_grade': self.get_privacy_grade(privacy_score),
            'risk_score': adjusted_risk_score, # Return adjusted risk
            'cookie_details': cookie_details,
            'third_party_cookies': third_party_count,
            'third_party_ratio': round(third_party_ratio, 1),
            'marketing_ratio': round(marketing_ratio, 1),
            'total_size_bytes': total_size,
            'total_size_readable': self.format_bytes(total_size),
            'tracker_companies': list(tracker_companies),
            'recommendations': self.generate_enhanced_recommendations(category_counts, privacy_score, third_party_ratio, list(tracker_companies)),
            'security_alerts': self.generate_security_alerts(cookie_details),
            'compliance_flags': self.generate_compliance_flags(cookie_details, category_counts)
        }


    def format_bytes(self, bytes_size):
        """Format bytes to human readable format"""
        if bytes_size < 1024:
            return f"{bytes_size} bytes"
        elif bytes_size < 1024 * 1024:
            return f"{bytes_size / 1024:.1f} KB"
        else:
            return f"{bytes_size / (1024 * 1024):.1f} MB"

    def get_risk_level(self, category):
        """Convert category to risk level with enhanced descriptions"""
        risk_mapping = {
            'ESSENTIAL': 'Low',
            'FUNCTIONAL': 'Medium',
            'ANALYTICAL': 'High',
            'MARKETING': 'Critical'
        }
        return risk_mapping.get(category, 'Unknown')

    def get_privacy_grade(self, score):
        """Enhanced privacy grading system"""
        if score >= 95: return 'A+'
        elif score >= 90: return 'A'
        elif score >= 85: return 'A-'
        elif score >= 80: return 'B+'
        elif score >= 75: return 'B'
        elif score >= 70: return 'B-'
        elif score >= 65: return 'C+'
        elif score >= 60: return 'C'
        elif score >= 55: return 'C-'
        elif score >= 50: return 'D'
        else: return 'F'

    def generate_enhanced_recommendations(self, category_counts, privacy_score, third_party_ratio, tracker_companies):
        """Generate detailed, actionable recommendations"""
        recommendations = []

        if category_counts.get('MARKETING', 0) > 1: # Reduced threshold for marketing rec
            recommendations.append({
                'priority': 'High',
                'category': 'Privacy Protection',
                'title': 'Multiple Marketing Cookies Detected',
                'description': f"Found {category_counts['MARKETING']} marketing cookies that may track you across websites.",
                'action': 'Review necessity and implement granular consent for marketing cookies',
                'impact': 'Significantly improves user privacy and consent compliance',
                'icon': 'üõ°Ô∏è'
            })

        if third_party_ratio > 30: # Reduced threshold for third-party rec
            recommendations.append({
                'priority': 'Medium',
                'category': 'Data Minimization',
                'title': 'High Third-Party Cookie Usage',
                'description': f"{third_party_ratio:.1f}% of cookies are from third-party domains.",
                'action': 'Reduce reliance on third-party tracking scripts or use server-side tracking',
                'impact': 'Minimizes cross-site tracking and external data sharing',
                'icon': 'üö´'
            })

        if category_counts.get('ANALYTICAL', 0) > 2: # Reduced threshold for analytical rec
            recommendations.append({
                'priority': 'Medium',
                'category': 'Behavioral Tracking',
                'title': 'Extensive Analytics Tracking',
                'description': f"Platform uses {category_counts['ANALYTICAL']} analytics cookies to monitor behavior.",
                'action': 'Evaluate data collected by analytics cookies and ensure data minimization',
                'impact': 'Reduces behavioral profiling and complies with data minimization principles',
                'icon': 'üìä'
            })

        if privacy_score < 75: # Adjusted threshold for overall privacy score
            recommendations.append({
                'priority': 'High',
                'category': 'Overall Privacy',
                'title': 'Privacy Score Needs Improvement',
                'description': f"Current privacy score is {privacy_score}/100, indicating significant privacy concerns.",
                'action': 'Implement recommendations across cookie management, consent, and data minimization',
                'impact': 'Leads to a more privacy-friendly platform and increased user trust',
                'icon': '‚ö†Ô∏è'
            })
        elif privacy_score >= 90:
             recommendations.append({
                'priority': 'Low',
                'category': 'Overall Privacy',
                'title': 'Excellent Privacy Practices',
                'description': f"Current privacy score is {privacy_score}/100, indicating strong privacy measures.",
                'action': 'Maintain current practices and monitor for future regulatory changes',
                'impact': 'Builds strong user loyalty and trust',
                'icon': '‚úÖ'
            })


        # Add recommendations based on identified tracker companies
        if 'Google' in tracker_companies:
             recommendations.append({
                'priority': 'Medium',
                'category': 'Third-Party Integrations',
                'title': 'Google Trackers Detected',
                'description': 'Using Google Analytics and/or DoubleClick cookies.',
                'action': 'Review Google\'s data processing terms and consider privacy-friendly alternatives',
                'impact': 'Reduces reliance on major advertising networks',
                'icon': 'üîç'
            })
        if 'Meta (Facebook)' in tracker_companies:
             recommendations.append({
                'priority': 'High',
                'category': 'Third-Party Integrations',
                'title': 'Meta (Facebook) Trackers Detected',
                'description': 'Using Facebook Pixel cookies.',
                'action': 'Ensure explicit consent for Facebook Pixel and review data sharing practices',
                'impact': 'Crucial for GDPR and CCPA compliance related to social media tracking',
                'icon': 'üëç'
            })

        return recommendations

    def generate_security_alerts(self, cookie_details):
        """Generate security-related alerts"""
        alerts = []

        # Check for insecure cookies (simplified simulation)
        # In a real scan, this would involve checking the 'Secure' flag
        simulated_insecure_count = sum(1 for c in cookie_details if c['category'] in ['ANALYTICAL', 'MARKETING'] and np.random.rand() < 0.1) # Simulate some insecure cookies

        if simulated_insecure_count > 0:
            alerts.append({
                'level': 'Warning',
                'title': 'Potential Insecure Cookie Transmission',
                'description': f'Detected {simulated_insecure_count} cookies that may lack the "Secure" flag, potentially transmitted over unencrypted connections.',
                'recommendation': 'Ensure all cookies, especially those with sensitive data or for tracking, are set with the "Secure" flag and that your platform strictly enforces HTTPS.'
            })

        # Check for long-lived tracking cookies (simulated based on purpose/category)
        simulated_long_lived_tracking = sum(1 for c in cookie_details if c['category'] in ['ANALYTICAL', 'MARKETING'] and np.random.rand() < 0.3) # Simulate some long-lived cookies

        if simulated_long_lived_tracking > 2:
            alerts.append({
                'level': 'Info',
                'title': 'Persistent Tracking Cookies Identified',
                'description': f'Found {simulated_long_lived_tracking} cookies used for analytics or marketing that appear to have long expiration periods.',
                'recommendation': 'Periodically review cookie expiration dates and minimize the lifespan of non-essential tracking cookies in accordance with data minimization principles.'
            })

        # Simulate check for cookies missing HTTPOnly flag (security risk for XSS)
        simulated_missing_httponly = sum(1 for c in cookie_details if c['category'] in ['ESSENTIAL', 'FUNCTIONAL'] and np.random.rand() < 0.05) # Simulate some missing HttpOnly

        if simulated_missing_httponly > 0:
             alerts.append({
                'level': 'Warning',
                'title': 'Potential Missing HttpOnly Flag',
                'description': f'Detected {simulated_missing_httponly} cookies that might be missing the "HttpOnly" flag, making them potentially accessible via JavaScript and vulnerable to XSS attacks.',
                'recommendation': 'Ensure all cookies, especially session and authentication cookies, are set with the "HttpOnly" flag to prevent client-side script access.'
            })

        return alerts


    def generate_compliance_flags(self, cookie_details, category_counts):
        """Generate compliance-related flags"""
        flags = []

        # GDPR compliance check
        if category_counts.get('MARKETING', 0) > 0 or category_counts.get('ANALYTICAL', 0) > 0:
            flags.append({
                'regulation': 'GDPR',
                'status': 'Requires Consent Mechanism',
                'description': 'Non-essential cookies (Marketing, Analytics) require explicit user consent under GDPR. Ensure a compliant cookie banner and consent management platform are in place.'
            })
        else:
             flags.append({
                'regulation': 'GDPR',
                'status': 'Likely Compliant (Cookies)',
                'description': 'Based on cookie categories detected, the platform likely adheres to GDPR principles regarding cookie consent, assuming other data processing is compliant.'
            })

        # CCPA compliance check
        if category_counts.get('MARKETING', 0) > 0 or category_counts.get('ANALYTICAL', 0) > 1:
             flags.append({
                'regulation': 'CCPA',
                'status': 'Requires Opt-Out Option',
                'description': 'Tracking cookies may be considered "sale" of personal information under CCPA. Ensure a clear "Do Not Sell My Personal Information" link and opt-out mechanism is available.'
            })
        else:
             flags.append({
                'regulation': 'CCPA',
                'status': 'Likely Compliant (Cookies)',
                'description': 'Based on cookie categories detected, the platform likely adheres to CCPA principles regarding cookie opt-out, assuming other data processing is compliant.'
            })


        # COPPA compliance check (for educational platforms dealing with minors)
        # This is a heuristic check based on cookie types commonly associated with tracking
        potentially_tracking_minors = sum(1 for c in cookie_details if c['category'] in ['ANALYTICAL', 'MARKETING'])
        if potentially_tracking_minors > 0: # Any analytical or marketing cookie is a potential COPPA issue if minors are users
            flags.append({
                'regulation': 'COPPA',
                'status': 'Potential Issue (if serving minors)',
                'description': 'Tracking cookies may violate COPPA if the platform serves children under 13. Verifiable parental consent is required before collecting personal info via cookies from children.'
            })
        else:
             flags.append({
                'regulation': 'COPPA',
                'status': 'Likely Compliant (Cookies for Minors)',
                'description': 'Based on cookie categories, the platform seems to minimize tracking relevant to COPPA, assuming no personal info is collected via essential/functional cookies from minors without consent.'
            })


        # FERPA compliance check (for US educational institutions/platforms)
        # This is a heuristic check focusing on cookies that might store educational records or link to them
        potentially_ferpa_relevant = sum(1 for c in cookie_details if c['category'] in ['FUNCTIONAL', 'ESSENTIAL'] and ('progress' in c['name'] or 'enrollment' in c['name'] or 'user-info' in c['name'] or 'session' in c['name']))

        if potentially_ferpa_relevant > 0:
             flags.append({
                'regulation': 'FERPA',
                'status': 'Requires Review',
                'description': 'Some functional/essential cookies might handle data related to student educational records. Ensure these comply with FERPA disclosure and security requirements.'
            })
        else:
             flags.append({
                'regulation': 'FERPA',
                'status': 'Likely Compliant (Cookies)',
                'description': 'Based on cookie names and categories, the platform seems less likely to store sensitive educational records in cookies, which aligns with FERPA principles.'
            })

        # DPDP compliance check (India)
        if category_counts.get('MARKETING', 0) > 0 or category_counts.get('ANALYTICAL', 0) > 0:
            flags.append({
                'regulation': 'DPDP',
                'status': 'Requires Consent and Data Minimization',
                'description': 'DPDP requires valid consent and adherence to data minimization. Ensure all collected data via cookies is necessary and consent is obtained for non-essential uses.'
            })
        else:
             flags.append({
                'regulation': 'DPDP',
                'status': 'Likely Compliant (Cookies)',
                'description': 'Based on cookie categories, the platform likely adheres to DPDP principles regarding consent and data minimization for cookies.'
            })


        return flags

# Initialize enhanced analysis engine
enhanced_cookie_engine = EnhancedCookieAnalysisEngine()
print("‚úÖ Enhanced Cookie Analysis Engine initialized!")
print("üîç Advanced privacy analysis capabilities loaded!")
````

````python
# Cell 4: Enhanced Compliance Checker with Global Regulations
# ==========================================================
class EnhancedComplianceChecker:
    def __init__(self):
        self.regulations = {
            'GDPR': {
                'name': 'General Data Protection Regulation',
                'region': 'European Union',
                'icon': 'üá™üá∫',
                'effective_date': '2018-05-25',
                'key_requirements': [
                    'Explicit consent for non-essential cookies',
                    'Clear and accessible privacy policy',
                    'Right to withdraw consent easily',
                    'Data minimization principle',
                    'Granular cookie controls',
                    'Privacy by design implementation'
                ],
                'penalties': 'Up to ‚Ç¨20M or 4% of global revenue',
                'student_focus': 'Special protection for minors under 16'
            },
            'CCPA': {
                'name': 'California Consumer Privacy Act',
                'region': 'California, USA',
                'icon': 'üá∫üá∏',
                'effective_date': '2020-01-01',
                'key_requirements': [
                    'Right to know about data collection',
                    'Right to delete personal information',
                    'Right to opt-out of sale of personal info',
                    'Clear privacy policy with categories',
                    '"Do Not Sell My Personal Information" link',
                    'Non-discrimination for privacy choices'
                ],
                'penalties': 'Up to $7,500 per violation',
                'student_focus': 'Enhanced protections for minors'
            },
            'DPDP': {
                'name': 'Digital Personal Data Protection Act',
                'region': 'India',
                'icon': 'üáÆüá≥',
                'effective_date': '2023-08-11',
                'key_requirements': [
                    'Valid consent for data processing',
                    'Data minimization practices',
                    'Purpose limitation compliance',
                    'Storage limitation adherence',
                    'Right to erasure mechanism',
                    'Verifiable parental consent for minors'
                ],
                'penalties': 'Up to ‚Çπ250 crores',
                'student_focus': 'Special consent requirements for under 18'
            },
            'COPPA': {
                'name': 'Children\'s Online Privacy Protection Act',
                'region': 'United States',
                'icon': 'üë∂',
                'effective_date': '2000-04-21',
                'key_requirements': [
                    'Verifiable parental consent for under 13',
                    'Clear privacy notice for children',
                    'Limited collection of personal information',
                    'No behavioral advertising to children',
                    'Safe harbor provisions compliance',
                    'Regular deletion of child data'
                ],
                'penalties': 'Up to $43,792 per violation',
                'student_focus': 'Primary focus on children under 13'
            },
            'FERPA': {
                'name': 'Family Educational Rights and Privacy Act',
                'region': 'United States',
                'icon': 'üéì',
                'effective_date': '1974-08-21',
                'key_requirements': [
                    'Consent for disclosure of education records',
                    'Right to access and review education records',
                    'Right to request amendment of records',
                    'Limited disclosure without consent',
                    'Annual notification of rights',
                    'Secure handling of student data'
                ],
                'penalties': 'Loss of federal education funding',
                'student_focus': 'Protects all student education records'
            }
        }

        self.compliance_weights = {
            'privacy_score': 0.3,
            'consent_mechanism': 0.25,
            'data_minimization': 0.2,
            'transparency': 0.15,
            'user_rights': 0.1
        }

    def check_enhanced_compliance(self, analysis_results, platform_category='EdTech'):
        """Enhanced compliance checking with detailed analysis"""
        compliance_results = {}

        for reg_code, regulation in self.regulations.items():
            # Calculate base compliance score
            base_score = self.calculate_detailed_compliance_score(reg_code, analysis_results, platform_category)

            # Get compliance status and recommendations
            status = self.get_detailed_compliance_status(base_score)
            recommendations = self.get_detailed_compliance_recommendations(reg_code, base_score, analysis_results)

            compliance_results[reg_code] = {
                'name': regulation['name'],
                'region': regulation['region'],
                'icon': regulation['icon'],
                'effective_date': regulation['effective_date'],
                'score': base_score,
                'status': status,
                'requirements': regulation['key_requirements'],
                'penalties': regulation['penalties'],
                'student_focus': regulation['student_focus'],
                'recommendations': recommendations,
                'risk_level': self.get_risk_level(base_score),
                'compliance_grade': self.get_compliance_grade(base_score)
            }

        # Add overall compliance summary
        overall_score = np.mean([result['score'] for result in compliance_results.values()])
        compliance_results['OVERALL'] = {
            'name': 'Overall Compliance Assessment',
            'score': round(overall_score, 1),
            'status': self.get_detailed_compliance_status(overall_score),
            'risk_level': self.get_risk_level(overall_score),
            'compliance_grade': self.get_compliance_grade(overall_score)
        }

        return compliance_results

    def calculate_detailed_compliance_score(self, reg_code, analysis_results, platform_category):
        """Calculate detailed compliance score with multiple factors"""
        base_score = 50  # Starting baseline

        # Privacy score impact (30% weight)
        privacy_score = analysis_results['privacy_score']
        privacy_impact = (privacy_score - 50) * self.compliance_weights['privacy_score']

        # Consent mechanism assessment (25% weight)
        marketing_cookies = analysis_results['category_counts'].get('MARKETING', 0)
        consent_score = max(0, 100 - (marketing_cookies * 15))
        consent_impact = (consent_score - 50) * self.compliance_weights['consent_mechanism']

        # Data minimization assessment (20% weight)
        total_cookies = analysis_results['total_cookies']
        minimization_score = max(0, 100 - (total_cookies * 5))
        minimization_impact = (minimization_score - 50) * self.compliance_weights['data_minimization']

        # Transparency assessment (15% weight)
        third_party_ratio = analysis_results.get('third_party_ratio', 0)
        transparency_score = max(0, 100 - third_party_ratio)
        transparency_impact = (transparency_score - 50) * self.compliance_weights['transparency']

        # User rights assessment (10% weight)
        functional_cookies = analysis_results['category_counts'].get('FUNCTIONAL', 0)
        rights_score = min(100, 50 + (functional_cookies * 10))
        rights_impact = (rights_score - 50) * self.compliance_weights['user_rights']

        # Regulation-specific adjustments
        regulation_bonus = self.get_regulation_specific_bonus(reg_code, analysis_results, platform_category)

        final_score = base_score + privacy_impact + consent_impact + minimization_impact + transparency_impact + rights_impact + regulation_bonus

        return max(0, min(100, round(final_score, 1)))

    def get_regulation_specific_bonus(self, reg_code, analysis_results, platform_category):
        """Get regulation-specific bonus/penalty adjustments"""
        bonus = 0

        if reg_code == 'GDPR':
            # GDPR is stricter on marketing cookies
            if analysis_results['category_counts'].get('MARKETING', 0) == 0:
                bonus += 10
            elif analysis_results['category_counts'].get('MARKETING', 0) > 3:
                bonus -= 15

        elif reg_code == 'COPPA' or reg_code == 'FERPA':
            # Educational platforms have stricter requirements
            if platform_category in ['K-12 Education', 'Educational']:
                if analysis_results['category_counts'].get('ANALYTICAL', 0) > 2:
                    bonus -= 20  # Heavy penalty for extensive tracking in education
                if analysis_results['category_counts'].get('MARKETING', 0) > 0:
                    bonus -= 25  # Very heavy penalty for marketing to students

        elif reg_code == 'DPDP':
            # DPDP focuses on data minimization
            if analysis_results['total_cookies'] <= 5:
                bonus += 15
            elif analysis_results['total_cookies'] > 10:
                bonus -= 10

        return bonus

    def get_detailed_compliance_status(self, score):
        """Get detailed compliance status"""
        if score >= 90: return 'üü¢ Excellent Compliance'
        elif score >= 80: return 'üü° Good Compliance'
        elif score >= 70: return 'üü† Moderate Compliance'
        elif score >= 60: return 'üî¥ Poor Compliance'
        elif score >= 40: return '‚ùå Critical Issues'
        else: return 'üö® Severe Non-Compliance'

    def get_risk_level(self, score):
        """Get risk level based on compliance score"""
        if score >= 80: return 'Low Risk'
        elif score >= 60: return 'Medium Risk'
        elif score >= 40: return 'High Risk'
        else: return 'Critical Risk'

    def get_compliance_grade(self, score):
        """Get letter grade for compliance"""
        if score >= 95: return 'A+'
        elif score >= 90: return 'A'
        elif score >= 85: return 'B+'
        elif score >= 80: return 'B'
        elif score >= 75: return 'C+'
        elif score >= 70: return 'C'
        elif score >= 65: return 'D+'
        elif score >= 60: return 'D'
        else: return 'F'

    def get_detailed_compliance_recommendations(self, reg_code, score, analysis_results):
        """Get detailed, actionable compliance recommendations"""
        recommendations = []

        if reg_code == 'GDPR':
            if score < 80:
                recommendations.extend([
                    {
                        'priority': 'High',
                        'title': 'Implement Granular Cookie Consent',
                        'description': 'Add cookie banner with separate opt-ins for different cookie categories',
                        'timeline': '2-4 weeks',
                        'effort': 'Medium'
                    },
                    {
                        'priority': 'High',
                        'title': 'Review Data Processing Lawful Basis',
                        'description': 'Ensure legitimate interests assessment for analytics cookies',
                        'timeline': '1-2 weeks',
                        'effort': 'Low'
                    }
                ])

            if analysis_results['category_counts'].get('MARKETING', 0) > 0:
                recommendations.append({
                    'priority': 'Critical',
                    'title': 'Marketing Cookie Consent Required',
                    'description': 'All marketing cookies require explicit opt-in consent under GDPR',
                    'timeline': 'Immediate',
                    'effort': 'High'
                })

        elif reg_code == 'COPPA':
            if analysis_results['category_counts'].get('ANALYTICAL', 0) > 1:
                recommendations.append({
                    'priority': 'Critical',
                    'title': 'Remove Behavioral Tracking for Minors',
                    'description': 'COPPA prohibits behavioral tracking of children under 13',
                    'timeline': 'Immediate',
                    'effort': 'High'
                })

            recommendations.append({
                'priority': 'High',
                'title': 'Implement Age Verification',
                'description': 'Add age verification mechanism before cookie deployment',
                'timeline': '4-6 weeks',
                'effort': 'High'
            })

        elif reg_code == 'FERPA':
            recommendations.extend([
                {
                    'priority': 'High',
                    'title': 'Student Data Protection Audit',
                    'description': 'Ensure cookies don\'t inadvertently collect educational records',
                    'timeline': '2-3 weeks',
                    'effort': 'Medium'
                },
                {
                    'priority': 'Medium',
                    'title': 'Third-party Vendor Agreements',
                    'description': 'Review contracts with analytics and advertising providers',
                    'timeline': '3-4 weeks',
                    'effort': 'Medium'
                }
            ])

        return recommendations

# Initialize enhanced compliance checker
enhanced_compliance_checker = EnhancedComplianceChecker()
print("‚úÖ Enhanced Compliance Checker initialized!")
print("‚öñÔ∏è Global privacy regulations loaded!")
````

## Create flask app

### Subtask:
Create a new Flask application file (`app.py`).


**Reasoning**:
I will create a new Flask application file named `app.py`. Inside this file, I will import the necessary classes from the modified notebook script and the Flask library. Then, I will instantiate the Flask application and the core analysis engines to set up the basic structure of the web application.



````python
# app.py

from flask import Flask, request, jsonify, render_template_string
from your_modified_notebook import (
    EnhancedCookieAnalysisEngine,
    EnhancedComplianceChecker,
    AdvancedVisualizationEngine,
    CookieBeeUtilities
)

# Instantiate the Flask application
app = Flask(__name__)

# Instantiate the core analysis engines
analysis_engine = EnhancedCookieAnalysisEngine()
compliance_checker = EnhancedComplianceChecker()
viz_engine = AdvancedVisualizationEngine()
utilities = CookieBeeUtilities()

# Simple HTML form for user input
HOME_HTML = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>CookieBee Analysis</title>
    <style>
        body { font-family: sans-serif; margin: 2em; }
        input { font-size: 1.2em; padding: 0.5em; }
        button { font-size: 1.2em; padding: 0.5em; }
    </style>
</head>
<body>
    <h1>CookieBee - Analyze EdTech Privacy</h1>
    <form action="/analyze" method="post">
        <label for="url">Enter Website URL:</label><br><br>
        <input type="text" id="url" name="url" size="50" value="https://www.khanacademy.org">
        <button type="submit">Analyze</button>
    </form>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(HOME_HTML)

@app.route('/analyze', methods=['POST'])
def analyze():
    url = request.form.get('url')
    if not url:
        return jsonify({"error": "URL is required"}), 400

    if not url.startswith(('http://', 'https://')):
        url = 'https://' + url

    try:
        # Perform analysis
        scan_data = analysis_engine.simulate_advanced_cookie_scan(url)
        analysis_results = analysis_engine.analyze_cookies_advanced(scan_data)
        compliance_results = compliance_checker.check_enhanced_compliance(
            analysis_results, scan_data.get('platform_category', 'EdTech')
        )

        # Generate a basic summary to return as JSON
        summary = {
            "platform_info": {
                "name": scan_data.get('platform_name'),
                "category": scan_data.get('platform_category'),
                "url": url,
            },
            "privacy_analysis": {
                "total_cookies": analysis_results.get('total_cookies'),
                "privacy_score": analysis_results.get('privacy_score'),
                "privacy_grade": analysis_results.get('privacy_grade'),
                "third_party_cookies": analysis_results.get('third_party_cookies'),
                "category_counts": analysis_results.get('category_counts'),
            },
            "compliance_summary": {
                "overall_score": compliance_results.get('OVERALL', {}).get('score'),
                "risk_level": compliance_results.get('OVERALL', {}).get('risk_level'),
            }
        }
        return jsonify(summary)

    except Exception as e:
        return jsonify({"error": f"An error occurred: {str(e)}"}), 500

if __name__ == '__main__':
    app.run(debug=True)
````
